```python
# Feature Engineering, 
# extract features from existing features

df['job-title'].value_counts()
```

```python
# let's make a new feature job_status which is  going to store if the employe is current or former eomployee

job_status = []
designation = []
des_string = ''

for i in range(len(df['job-title'])):
    if "Current Employee" in df['job-title'][i]:
        job_status.append("Current Employee")
        des_string = df['job-title'][i][18:]
        designation.append(des_string)
        
    elif "Former Employee" in df['job-title'][i]:
        job_status.append("Former Employee")
        des_string = df['job-title'][i][17:]
        designation.append(des_string)
        
    
#print(job_status)

# merge the new feature to the datafram
df['job_status'] = job_status
df['job_title'] = designation
df.head()
```

```python
# convert the 'dates' feature to pandas date time format and then extract day, month and year as features
#df['dates'] = pd.to_datetime(df.dates , format='%b %d, %Y')


# make 3 new features, day_of_the_month, month, year and extract them from the dates feature whose values are in the format
# of string

post_day = []
post_month = []
post_year = []
temp = ''

for i in range(len(df['dates'])):
    temp = df['dates'][i][1:4]
    post_month.append(temp)

    temp = df['dates'][i][-8:-6]
    post_day.append(temp)    
    
    temp = df['dates'][i][-5:]
    post_year.append(temp)
       

# merge the new feature to the datafram
df['post_day'] = post_day
df['post_month'] = post_month
df['post_year'] = post_year
df.head()
```

```python
df['job_status'].value_counts()
```

```python
# if we are considering job_status as our Target for classification then surely it;s a class impabalance problem
# let's do undersampling to keep all the values equal for both the classes 
# chose 20000 values from both the classes randomly

# use resample method from scikit-learn
from sklearn.utils import resample

# Separate majority and minority classes for undersampling
df_Current_Employee  = df[df.job_status== 'Current Employee']
df_Former_Employee   = df[df.job_status == 'Former Employee']
    
df_Current_Employee = resample(df_Current_Employee, 
                                             replace=True,    # sample with replacement
                                             n_samples=20000,     # to match number of values in each class
                                             random_state=123) # reproducible results

df_Former_Employee   = resample(df_Former_Employee, 
                                             replace=True,    # sample with replacement
                                             n_samples=20000,     # to match number of values in each class
                                             random_state=123) # reproducible results

# Combine all the class with equal number of values
df_sampled = pd.concat([ df_Current_Employee , df_Former_Employee])
 

# check for the class balance
print(df_sampled['job_status'].value_counts())
# change df with sampled one
df= df_sampled
print("\n")
print(df.head(5))
```

```python
# work on the "none" values in the rows
df['location'] = df['location'].replace(['none'], 'Unknown')
```

```python
# as all the ratings have "none" as a value except overall-ratings, we need to replace it with the average rating value 2.5
# reason: 2.5 is not too high not too low, not influence the models in a significant way, that's what we want here
df['work-balance-stars'] = df['work-balance-stars'].replace(['none'], 2.5)
df['culture-values-stars'] = df['culture-values-stars'].replace(['none'], 2.5)
df['carrer-opportunities-stars'] = df['carrer-opportunities-stars'].replace(['none'], 2.5)
df['comp-benefit-stars'] = df['comp-benefit-stars'].replace(['none'], 2.5)
df['senior-mangemnet-stars'] = df['senior-mangemnet-stars'].replace(['none'], 2.5)
```

```python
# work on the none/blank/error values of the dates (day, month, year)
df.post_day.value_counts().sort_values()
```

```python
# replace the day of the month 0 and blank with 15
df['post_day'] = df['post_day'].replace([' 0' , ''], 15)
```

```python
# check if the values been replaced
#change the type of the post_day from string to int
df.post_day = df.post_day.astype(np.int64)

df.post_day.value_counts().sort_values()
```

```python
# Drop null values
df = df.dropna(how="any")
```

```python
# as there are no none values for month, we just need to work on the year
# replace 0000, None with 2009, as there are only 4 such values, so it doesnt matter much where to put the values

df['post_year'] = df['post_year'].replace([' 0000' , 'None'], '2009')
```
